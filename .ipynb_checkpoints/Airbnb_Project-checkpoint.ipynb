```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from scipy.stats import ttest_ind
from scipy.stats import f_oneway, chi2_contingency
```

```python
class EDA:
    #read data
    def __init__(self,fileNmae):
        self.fileNmae=fileNmae
        self.data=None
        extension = fileNmae.split('.')[-1].lower()
        try:
            if extension == 'csv':
                self.data=pd.DataFrame(pd.read_csv(fileNmae))
            elif extension == 'xls' or extension == 'xlsx':
                self.data=pd.DataFrame(pd.read_excel(fileNmae))
            elif extension == 'json':
                self.data=pd.DataFrame(pd.read_json(fileNmae))
            elif extension == 'sql':
                self.data=pd.DataFrame(pd.read_sql(fileNmae))
            elif extension == 'html':
                self.data=pd.DataFrame(pd.read_html(fileNmae))
            elif extension == 'feather':
                self.data=pd.DataFrame(pd.read_feather(fileNmae))
            elif extension == 'spss':
                self.data=pd.DataFrame(pd.read_spss(fileNmae))
            else:
                return 'Sorry, I can only read files with the extension: .csv, .xls or .xlsx, .json, .sql, .html, .feather, .spss'
        except FileNotFoundError :
            print('The file does not exist or you made a mistake in the file name')
    def analyzing_KeyFactors_Influencing_Demand(self):
        data = self.data
        # basic information about the dataset
        title = "Basic Information"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title) + "\n" + title + "\n" + "#" * len(title))
        print(data.info())

        title = "Analyzing Key Factors Influencing Demand"
        title2 = "The Key Factors are: Price ,Availability , Location, Property Type, Neighborhood, Amenities, Review Scores, Number of Bedrooms, Bathrooms, and Beds"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Price Distribution
        plt.figure(figsize=(10, 6))
        sns.histplot(data['price'], bins=30, kde=True, color='skyblue')
        plt.title('Distribution of Prices')
        plt.xlabel('Price')
        plt.ylabel('Frequency')
        plt.show()
        title = "Price Distribution"
        title2 = "واضح من الكيرف المنحني نحو اليسار ان أغلب الحجوزات على السعر الارخص اي ان السعر عامل مؤثر جدا على الطلب"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Availability Trends
        availability_trends = data[['calendar_last_scraped', 'availability_30', 'availability_60', 'availability_90', 'availability_365']].groupby('calendar_last_scraped').mean().reset_index()

        plt.figure(figsize=(12, 8))
        plt.plot(availability_trends['calendar_last_scraped'], availability_trends['availability_30'], label='Availability (30 days)', marker='o', color='blue', linewidth=2)
        plt.plot(availability_trends['calendar_last_scraped'], availability_trends['availability_60'], label='Availability (60 days)', marker='s', color='green', linewidth=2)
        plt.plot(availability_trends['calendar_last_scraped'], availability_trends['availability_90'], label='Availability (90 days)', marker='^', color='orange', linewidth=2)
        plt.plot(availability_trends['calendar_last_scraped'], availability_trends['availability_365'], label='Availability (365 days)', marker='x', color='red', linewidth=2)
        plt.title('Availability Trends Over Different Time Frames')
        plt.xlabel('Month')
        plt.ylabel('Average Availability (in days)')
        plt.xticks(rotation=45)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
        title = "Availability Trends"
        title2 = "نستنتج من الشكل اعلاه ان معدل قيم الحجز الشهرية و حتى السنوية ادنى من المعدل الطبيعي اي ان بعض العملاء قد غادروا"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Neighborhood Distribution
        plt.figure(figsize=(12, 8))
        sns.countplot(y='neighbourhood_cleansed', data=data, order=data['neighbourhood_cleansed'].value_counts().index, palette='viridis')
        plt.title('Distribution of Listings by Neighborhood')
        plt.xlabel('Number of Listings')
        plt.ylabel('neighbourhood_cleansed')
        plt.show()
        title = "Neighborhood Distribution"
        title2 = "الحي ايضا يؤثر على الطلب كما في الشكل اعلاه الحي السادس هوة الاكثر اختيار و يليه الثاني و هكذا "
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Property Type Distribution
        plt.figure(figsize=(8, 6))
        sns.countplot(y='property_type', data=data, order=data['property_type'].value_counts().index, palette='viridis')
        plt.title('Distribution of Listings by Property Type')
        plt.xlabel('Property Type')
        plt.ylabel('Number of Listings')
        plt.show()
        title = "Property Type Distribution"
        title2 = "كذالك نوع العقار يعتبر عامل مؤثر ايضا كما موضح في المخطط اعلاه"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Amenities
        amenities_count = data['amenities'].str.split(',').apply(len)
        plt.figure(figsize=(10, 6))
        sns.histplot(amenities_count, bins=30, kde=False, color='orange')
        plt.title('Distribution of Amenities')
        plt.xlabel('Number of Amenities')
        plt.ylabel('Frequency')
        plt.show()
        title = "Analyze Amenities"
        title2 = "و من البديهي ان تكون وسائل الراحة اكثر ما يهم العميل"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Review Scores
        review_scores_cols = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value']
        plt.figure(figsize=(12, 8))
        for col in review_scores_cols:
            sns.histplot(data[col], bins=30, kde=False, alpha=0.5, label=col)
        plt.title('Distribution of Review Scores')
        plt.xlabel('Review Score')
        plt.ylabel('Frequency')
        plt.legend()
        plt.show()
        title = "Review Scores"
        title2 = "ايضا كل تفاصيل التقييم و المراجعة يعتبر عامل مهم و مؤثر على الطلب"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Analyze Number of Bedrooms, Bathrooms, and Beds
        title = "Number of Bedrooms, Bathrooms, and Beds Analyze"
        title2 = "من الرسوم البيانية ادناه نستنتج ان اغلب العملاء يفضلون سرير و غرفة نوم و حمام واحد فقط"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))
        plt.figure(figsize=(12, 6))
        sns.countplot(x='bedrooms', data=data, palette='muted')
        plt.title('Distribution of Bedrooms')
        plt.xlabel('Number of Bedrooms')
        plt.ylabel('Number of Listings')
        plt.show()

        plt.figure(figsize=(12, 6))
        sns.countplot(x='bathrooms', data=data, palette='muted')
        plt.title('Distribution of Bathrooms')
        plt.xlabel('Number of Bathrooms')
        plt.ylabel('Number of Listings')
        plt.show()

        plt.figure(figsize=(12, 6))
        sns.countplot(x='beds', data=data, palette='muted')
        plt.title('Distribution of Beds')
        plt.xlabel('Number of Beds')
        plt.ylabel('Number of Listings')
        plt.show()


    def cleanData(self):
        title = "Clean data from missing value and outliers"
        title2 = "Data are cleaned"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))
    
        # Handling missing values
        numerical_cols = self.data.select_dtypes(include=[np.number]).columns
        categorical_cols = self.data.select_dtypes(include=['object']).columns
    
        self.data[numerical_cols] = self.data[numerical_cols].fillna(0)
        self.data[categorical_cols] = self.data[categorical_cols].fillna('null')
    
        # Removing outliers
        outlier_indices = set()
        for col in numerical_cols:
            Q1 = self.data[col].quantile(0.25)
            Q3 = self.data[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 3 * IQR 
            upper_bound = Q3 + 3 * IQR
            outlier_indices.update(self.data[(self.data[col] < lower_bound) | (self.data[col] > upper_bound)].index) 
    
        self.data = self.data.drop(outlier_indices)



    def calculateDemand(self):
        title = "Calculate High Demaend for each host"
        title2 = "من التحليل الاولي للبيانات وجدنا بعض القيم مرتفعة عند اغلب المضيفين و وفق هذه القيم سنقوم بحساب مقدار الطلب الاعلى"
        print("=" * 20 + title + "=" * 20)
        print("#" * len(title2) + "\n" + title2 + "\n" + "#" * len(title2))

        # Normalize numerical features
        numerical_features = ['availability_30', 'availability_60', 'availability_90', 'availability_365', 'bedrooms', 'bathrooms', 'beds']
        scaler = MinMaxScaler()
        self.data[numerical_features] = scaler.fit_transform(self.data[numerical_features])

        # weights for each feature
        weights = {
            'price': 0.2,
            'availability_30': 0.1,
            'availability_60': 0.1,
            'availability_90': 0.1,
            'availability_365': 0.1,
            'neighbourhood_cleansed': 0.05,
            'property_type': 0.05,
            'bedrooms': 0.05,
            'bathrooms': 0.05,
            'beds': 0.05,
            'amenities': 0.05,
            'review_scores_rating': 0.05,
            'review_scores_accuracy': 0.05,
            'review_scores_cleanliness': 0.05,
            'review_scores_checkin': 0.05,
            'review_scores_communication': 0.05,
            'review_scores_location': 0.05,
            'review_scores_value': 0.05
        }

        # Calculate demand score
        demand_score = sum(self.data[feature] * weights[feature] for feature in weights.keys() if feature in self.data.columns and self.data[feature].dtype in ['int64', 'float64'])

        # scale demand scores to a desired range
        min_demand_score = demand_score.min()
        max_demand_score = demand_score.max()
        demand_score_scaled = (demand_score - min_demand_score) / (max_demand_score - min_demand_score) * 100

        # Convert demand score to boolean column
        self.data['demand_high'] = (demand_score_scaled > 50).astype(int)

        # Print the demand scores
        print(self.data[['demand_high']])

    def AdvancedDataVisualization(self):
        title1 = " حساب تكرار او عدد الطلبات "
        print("=" * 20 + title1 + "=" * 20)
        print(self.data['demand_high'].value_counts())
        sns.countplot(x='demand_high',data=self.data)
        plt.show()
        
        title2 = " حساب نسبة السعر بالنسبة لعمود الطلب لان السعر مؤثر على الطلب "
        print("=" * 20 + title2 + "=" * 20)
        fig = px.pie(self.data, values=pd.to_numeric(self.data['price'].str.replace('[^\d.]', '', regex=True)), names='demand_high')
        fig.show()

        title2 = " عرض جدول الترابط بين الاعمدة للبيانات "
        print("=" * 20 + title2 + "=" * 20)
        numeric_data = self.data.select_dtypes(include=[np.number])
        plt.figure(figsize=(10,10))
        sns.heatmap(numeric_data.corr(),annot =True)
        plt.show()

        title2 = " رسم بياني توضيحي لكل الاعمدة المترابطة "
        print("=" * 20 + title2 + "=" * 20)
        features =['availability_30','availability_365', 'bedrooms', 'bathrooms','price','property_type','amenities','review_scores_rating','review_scores_location']
        sns.pairplot(self.data[features])
        plt.show()

        title2 = " تمييز قيمة ال0 و 1 المؤثرة على الطلب من اعمدة السعر و الخدمة المتاحة لمدة شهر "
        print("=" * 20 + title2 + "=" * 20)
        newData=self.data
        newData['price']=pd.to_numeric(self.data['price'].str.replace('[^\d.]', '', regex=True))
        sns.lmplot(x='price',y='availability_30',data=newData,hue='demand_high',fit_reg=False)
        plt.show()

        title2 = " فصل البيانات عن بعضها وفق عمود الطلب "
        print("=" * 20 + title2 + "=" * 20)
        x=self.data
        x['host_is_superhost']=x['host_is_superhost'].map({'f':0,'t':1})
        x['host_has_profile_pic']=x['host_has_profile_pic'].map({'f':0,'t':1})
        x['host_identity_verified']=x['host_identity_verified'].map({'f':0,'t':1})
        x['has_availability']=x['has_availability'].map({'f':0,'t':1})
        x['instant_bookable']=x['instant_bookable'].map({'f':0,'t':1})
        numeric_data = x.select_dtypes(include=[np.number])
        numeric_data_scl=StandardScaler().fit_transform(numeric_data)
        tsne=TSNE()
        numeric_data_Red=tsne.fit_transform(numeric_data_scl)
        plt.scatter(numeric_data_Red[:,0],numeric_data_Red[:,1],c=self.data['demand_high'].map({0:'orange',1:'blue'}),alpha=0.5)
        plt.show()
        title2 = " فصل البيانات عن بعضها وفق عمود التحقق من هوية المضيف "
        print("=" * 20 + title2 + "=" * 20)
        plt.scatter(numeric_data_Red[:,0],numeric_data_Red[:,1],c=self.data['host_identity_verified'].map({0:'orange',1:'blue'}),alpha=0.5)
        plt.show()

    def conductStatisticalAnalysis(self):
        # Convert 'price' column to numeric
        self.data['price'] = pd.to_numeric(self.data['price'], errors='coerce')

        # Drop rows with missing values in 'price' column
        self.data.dropna(subset=['price'], inplace=True)
        x=self.data
        x['host_is_superhost']=x['host_is_superhost'].map({'f':0,'t':1})
        x['host_has_profile_pic']=x['host_has_profile_pic'].map({'f':0,'t':1})
        x['host_identity_verified']=x['host_identity_verified'].map({'f':0,'t':1})
        x['has_availability']=x['has_availability'].map({'f':0,'t':1})
        x['instant_bookable']=x['instant_bookable'].map({'f':0,'t':1})
        numeric_data = x.select_dtypes(include=[np.number])

        # Correlation Analysis
        title = "Correlation Analysis"
        print("=" * 20 + title + "=" * 20)
        correlation_matrix = numeric_data.corr()
        print(correlation_matrix)

        # Hypothesis Testing
        title = "Hypothesis Testing"
        print("=" * 20 + title + "=" * 20)
        # Example: Compare average prices between different property types
        property_type_groups = self.data.groupby('property_type')['price']
        for property_type, group_data in property_type_groups:
            print(f"Average price for {property_type}: {group_data.mean()}")

        # Example: Perform t-test between two property types
        property_type_1 = self.data[self.data['property_type'] == 'Entire rental unit']['price']
        property_type_2 = self.data[self.data['property_type'] == 'Private room in home']['price']
        t_statistic, p_value = ttest_ind(property_type_1, property_type_2)
        print(f"T-statistic: {t_statistic}, p-value: {p_value}")



mydata=EDA('listings.csv')
mydata.analyzing_KeyFactors_Influencing_Demand()
mydata.cleanData()
mydata.calculateDemand()
mydata.AdvancedDataVisualization()
mydata.conductStatisticalAnalysis()
```
